```python
import json
import os

original_config_path = "/content/drive/MyDrive/phi3-merged/config.json"
with open(original_config_path, "r") as f:
    original_config = json.load(f)

mlc_config = {
    **original_config,
    "model_type": "phi3",
    "quantization": "q4f16_1",
    "model_lib": "phi3-webgpu",
    "conv_template": "phi-3",
    "context_window_size": original_config.get("max_position_embeddings", 4096),
    "prefill_chunk_size": 2048,
}

config_path = os.path.join(OUTPUT_DIR, "mlc-chat-config.json")
with open(config_path, "w") as f:
    json.dump(mlc_config, f, indent=2)

print(f"MLC configuration saved to: {config_path}")
!python -m mlc_llm compile \
  --device webgpu \
  --opt O3 \
  "/content/drive/MyDrive/phi3-mlc-output" \
  -o "/content/drive/MyDrive/phi3-mlc-output/phi3-webgpu.wasm"
```
